```python
#request each related link
links=[str(n) for n in post_links]
num=0
for link in post_links:
    comment_source=requests.get('https://www.ptt.cc/bbs/Gossiping/M.'+links[num]+'.html',cookies={'over18':'1'})
    num+=1
    #get the content with beautifulsoup
    comment_html=bs.BeautifulSoup(comment_source.content,'html.parser')
    #find itmes
  
    for comment_text in comment_html.find_all('span',class_='f3 push-content'):
        comment_texts.append(comment_text.text.strip())

    for comment_tage in comment_html.find_all('span',class_='f1 hl push-tag'):
        comment_tages.append(comment_tage.text.strip())

    for comment_time in comment_html.find_all('span',class_='push-ipdatetime'):
        comment_times.append(comment_time.text[-11:].strip())

    for comment_user in comment_html.find_all('span',class_='f3 hl push-userid'):
        comment_users.append(comment_user.text.strip())
    
#loading time
sleep(randint(1,2))
```
